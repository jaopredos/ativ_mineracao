# **ESTUDO DE CASO PARA MATÉRIA DE MINERAÇÃO DE DADOS**

```python
import pandas as pd

df = pd.read_csv('UCMF_raw.csv')
df.head()
```

Tamanho do conjunto de dados é:

```python
len(df)
```

## Retirando a coluna *ID*

```python
df_copy = df.copy()
df_copy.drop(columns=['ID'], inplace=True)
df_copy.head()
```

## Análise de dados faltantes nas colunas

### Peso

```python
print("Informações gerais sobre a coluna 'Peso':")
print(df_copy['Peso'].describe())

print("\n"+"-"*20)

print("\nNúmero de valores faltantes na coluna 'Peso':")
print(df_copy['Peso'].isna().sum())
```

Temos por volta de 2.5% dos dados dessa coluna faltante, não é muito preocupante. Preocupante é peso negativo e outros dados estranhos.

### Altura

```python
print("Informações gerais sobre a coluna 'Altura':")
print(df_copy['Altura'].describe())

print("\n"+"-"*20)

print("\nNúmero de valores faltantes na coluna 'Altura':")
print(df_copy['Altura'].isna().sum())
```

Não temos dados faltantes, isso é muito bom. Nem alturas negativas e dados estranhos nessa primeira análise.

### IMC

```python
print("Informações gerais sobre a coluna 'IMC':")
print(df_copy['IMC'].describe())

print("\n"+"-"*20)

print("\nNúmero de valores faltantes na coluna 'IMC':")
print(df_copy['IMC'].isna().sum())
```

Nessa coluna, possuímos 36.72% dos dados faltantes, isso já começa a ficar preocupante.

### Atendimento

```python
print("\nNúmero de valores faltantes na coluna 'Atendimento':")
print(df_copy['Atendimento'].isna().sum())
```

Nessa coluna, temos cerca de 7.6% de dados faltantes. Não vejo isso como problema, dado que são poucos dados faltantes e não vejo muita importância para essa coluna (data da visita do paciente).

```python
print("\nNúmero de valores faltantes na coluna 'DN':")
print(df_copy['DN'].isna().sum())
```
