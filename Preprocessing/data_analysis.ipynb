## Análise da relação dos atributos com as labels

```python
target_col = 'NORMAL X ANORMAL'
mapeamento_alvo = {
    'Anormal': 'Anormal',
    'anormal': 'Anormal',
    'Normal': 'Normal',
    'Normais': 'Normal'
}
df[target_col] = df[target_col].map(mapeamento_alvo)
df[target_col].value_counts()
```

```python
cat_cols = [col for col in df.select_dtypes(include='object').columns 
            if col != target_col and df[col].nunique() < 100]

print(f"Analisando as seguintes variáveis: {cat_cols}\n")


for col in cat_cols:
    print(f"--- Relação: {col} vs {target_col} ---")
    
    crosstab = pd.crosstab(df[col], df[target_col], dropna=False)
    
    crosstab['Total'] = crosstab.sum(axis=1)
    
    if 'Anormal' in crosstab.columns:
        crosstab['% Anormal'] = (crosstab['Anormal'] / crosstab['Total'] * 100).round(1)
    
    print(crosstab)
    print("\n" + "="*50 + "\n")
```
