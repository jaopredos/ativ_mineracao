# **PRÉ-PROCESSAMENTO DOS DADOS**

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\João Pedro\Documents\UFG\MD\estudo_caso\data\UCMF_raw.csv')
df.head()
```

## Retirando a coluna *ID* e *HDA2*

```python
df_copy = df.copy()
df_copy.drop(columns=['ID', 'HDA2'], inplace=True)
df_copy.head()
```

## Tratando a coluna *Idade*

```python
df_copy['DN'] = pd.to_datetime(df_copy['DN'], errors='coerce')
print(df_copy['DN'].isna().sum())

df_copy['Atendimento'] = pd.to_datetime(df_copy['Atendimento'], errors='coerce')
print(df_copy['Atendimento'].isna().sum())
```

```python
df_copy[df_copy['Atendimento'] >= df_copy['DN']]
```

## Tratando coluna *Sexo*

```python
dict_sexo = {'Masculino': 'M', 
             'Feminino': 'F',
             'masculino': 'M',
             'M': 'M',
             'F': 'F',
             'Indeterminado': 'I',}

df_copy['SEXO'] = df_copy['SEXO'].map(dict_sexo)
df_copy['SEXO'].value_counts()
```

```python
df_copy[df_copy['SEXO'].isna()]
```

```python
df_copy['SEXO'] = df_copy['SEXO'].map({'F': 0, 'M': 1, 'I': 2})
df_copy['SEXO'].value_counts()
```
